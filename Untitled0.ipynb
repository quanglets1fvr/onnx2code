```python
import torch
import torch.nn as nn

# Set the parameters
input_shape = (8, 8, 4)  # (Height, Width, Channels)
kernel_shape = (2, 4, 3, 3)
bias_shape = (1, 2, 1, 1)  # (Out_channels, In_channels, Kernel_height, Kernel_width)
padding = 'same'
stride = 1
dilation = 1
groups = 1

# Create the input tensor with shape (Batch_size, Channels, Height, Width)
input_tensor = torch.randint(0, 256, (1, input_shape[2], input_shape[0], input_shape[1]), dtype=torch.uint8)
bias_tensor = torch.randint(-20000, 20000 , bias_shape, dtype=torch.int32)

# Convert input tensor to float32 for convolution operation
input_tensor = input_tensor.float()

# Define the convolutional layer
conv_layer = nn.Conv2d(
    in_channels=kernel_shape[1],
    out_channels=kernel_shape[0],
    kernel_size=(kernel_shape[2], kernel_shape[3]),
    stride=stride,
    padding=padding,
    dilation=dilation,
    groups=groups,
    bias=False
)

# Initialize kernel weights with random uint8 values and convert to float32
kernel_weights = torch.randint(0, 256, kernel_shape, dtype=torch.uint8).float()
conv_layer.weight.data = kernel_weights

# Perform the convolution
output_tensor = conv_layer(input_tensor) + bias_tensor

# Print the output tensor
print("Output Tensor:")
print(output_tensor)
```
